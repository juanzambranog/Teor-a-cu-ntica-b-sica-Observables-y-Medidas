## Simule el primer sistema cuántico descrito en la sección 4.1.

```python
import numpy as np
class SistemaCuantico:
    def __init__(self, num_posiciones, vector_ket):
        self.num_posiciones = num_posiciones
        self.vector_ket = vector_ket

    def calcular_probabilidad(self, posicion):
        if posicion < 0 or posicion >= self.num_posiciones:
            raise ValueError("Posición fuera de rango")
        return abs(self.vector_ket[posicion])**2

    def calcular_probabilidad_transicion(self, vector_ket_final):
        if len(vector_ket_final) != self.num_posiciones:
            raise ValueError("Vector ket final tiene longitud incorrecta")
        producto_escalar = sum(a * b.conjugate() for a, b in zip(self.vector_ket, vector_ket_final))
        return abs(producto_escalar)**2
```

```python
import numpy as np
def calcular_amplitud_transicion(vector_inicial, vector_final):
    # Normalizar los vectores
    vector_inicial = vector_inicial / np.linalg.norm(vector_inicial)
    vector_final = vector_final / np.linalg.norm(vector_final)

    # Calcular la amplitud de transición
    amplitud_transicion = np.abs(np.dot(vector_inicial, vector_final))

    return amplitud_transicion
```
